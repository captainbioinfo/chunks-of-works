```r
# Load required library
library(dplyr)

# Read the CSV file
df <- read.csv("diversity_plot_data_cleaned.csv", check.names = FALSE)

# Combine duplicate genus rows by summing values across columns
df_merged <- df %>%
  group_by(GeneId) %>%
  summarise(across(.cols = everything(), .fns = sum), .groups = "drop")

# Save the result
write.csv(df_merged, "diversity_plot_data_cleaned2.csv", row.names = FALSE)
```
